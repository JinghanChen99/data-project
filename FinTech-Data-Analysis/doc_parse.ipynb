```
# web scraping
import requests
from bs4 import BeautifulSoup
import pprint
from time import sleep

# data manipulation
import pandas as pd
import numpy as np

# natural langauge processing 
import spacy
from collections import Counter
```

```
# pip3 install https://github.com/explosion/spacy-models/releases/download/en_core_web_sm-2.2.0/en_core_web_sm-2.2.0.tar.gz
nlp = spacy.load('en_core_web_sm')
```

```
tech_comp = pd.read_csv('tech_comp.csv')
tech_comp.head()
```

```
fin_keywords = ['app',
 'banking',
 'card reader',
 'eCommerce',
 'electronic',
 'finance',
 'financial',
 'financial data',
 'financial service',
 'insurance',
 'invest',
 'loans',
 'money transfer',
 'mortgage',
 'payment',
 'payments networks',
 'transaction',
 'transaction processing',
 'wallet']
```

```
tech_comp['S1-txt'] = tech_comp['SEC_url'].str[:-10] + ".txt"
tech_comp.head()
```

```
paypal = tech_comp[tech_comp['Company Name'] == 'PAYPAL INC']
paypal
```

```
paypal['S1-txt'].values[0]
```

```
r = requests.get(paypal['S1-txt'].values[0], headers={'User-Agent': 'Jinghan Chen  jchen911@gatech.edu'})
r.raise_for_status()
text_data = r.content 
sleep(0.1)
```

```
type(text_data)
```

```
text_data
```

```
nlp.max_length = 5000000
complete_doc = nlp(str(text_data), disable = ['ner', 'parser'])
# Remove stop words and punctuation symbols
words = [token.text for token in complete_doc if not token.is_stop and not token.is_punct]
word_freq = Counter(words)
# 5 commonly occurring words with their frequencies
common_words = word_freq.most_common()
# common_words
```

```
for tup in common_words:
    if tup[0] == 'payment':
        print(tup[1])
```

```
first, second, third, fourth = np.array_split(tech_comp, 4)
```

```
aDict = {}
for ind in first.index:
     aDict[first['Company Name'][ind]] = first['S1-txt'][ind]
# aDict
```

```
filtered_list = []
for key, value in aDict.items():
    r = requests.get(value, headers={'User-Agent': 'Jinghan Chen  jchen911@gatech.edu'})
    r.raise_for_status()
    text_data = r.content 
    sleep(0.1)

    nlp.max_length = 20000000
    complete_doc = nlp(str(text_data), disable = ['ner', 'parser'])
    # Remove stop words and punctuation symbols
    words = [token.text for token in complete_doc if not token.is_stop and not token.is_punct]
    word_freq = Counter(words)
    # frequency of all the words 
    common_words = word_freq.most_common()

    for tup in common_words:
        if (tup[0] in fin_keywords) & (tup[1] >= 10):
            # print(tup)
            if key not in filtered_list:
                filtered_list.append(key)
filtered_list
```

```
fintech = tech_comp[tech_comp['Company Name'].isin(filtered_list)]
fintech.shape
```
